Часто в приложениях встречаются задачи на работу с временными рядами. Практически всегда измерения заданы за различные участки времени. Перед исследователями встаёт задача: каким образом выравнять длины временных интервалов так, чтобы все временные ряды были одинаковой длины и образовывали матрицу. 

Данную задачу можно эффективно решить с помощью одного решающего дерева (НЕ случайного леса). Одно решающее дерево позволяет переобучиться и максимально точно передать те данные, хранит исходный сигнал. Таким образом, мы легко получаем все промежуточные значения. 

Данная задача называется **интерполяцией** и может быть решена одним из стандартных пакетов.

При этом, мы используем минимум пользовательского кода, который крайне просто пишется и читается. Здесь минимум граничных условий и практически нет нюансов. 

Важно отметить, что иногда требуется сжимать данные. Это легко корректируется с помощью задания числа отсчётов

```python
import json

from matplotlib import pyplot as plt
import numpy as np

import tseries
```

Прочитаем все временные ряды

```python
with open('data/ts.json') as fd:
    time_serieses = json.load(fd)
```

```python
index = 0
plt.figure()
plt.title(f'Source time series. Index: {index}')
plt.grid()
plt.plot(time_serieses[index]);
```

```python
index = 1
plt.figure()
plt.title(f'Source time series. Index: {index}')
plt.grid()
plt.plot(time_serieses[index]);
```

Построим апроксимацию временных рядов с помощью функционалов dtr (**DecisionTreeRegressor**) для заранее заданного чила точек (как меньше, чем в исходном наборе, так и большего):

```python
functionals = tseries.get_functionals(time_serieses)
```

Посмотрим, результат интерполяции:

```python
index = 0
interval = np.arange(0, 30)
plt.figure()
plt.title('Interpolation time series')
plt.plot(functionals[index].predict(tseries.any_to_01(interval)))
plt.grid()
plt.figure()
plt.title('Source time series')
plt.plot(time_serieses[index])
plt.grid()
```
